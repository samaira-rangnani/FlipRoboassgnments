Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
3. Then click the searchbutton.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
#webdriver ready
driver=webdriver.Chrome()
```

```python
#getting the website
driver.get('https://www.shine.com/')
```

```python
#finding the element
designation=driver.find_element(By.XPATH,'/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[1]/div/input')
designation.send_keys('Data Analyst')
```

```python
##finding the elemeent
location=driver.find_element(By.XPATH,'/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input')
location.send_keys('Banglore')
```

```python
#clicking on serach button
button=driver.find_element(By.XPATH,'/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button')
button.click()
time.sleep(2)
```

```python
#empty list to append desired data
job_title=[]
job_location=[] 
company_name=[] 
experience_required=[]
```

```python
###SCRAPPING JOB TIILES

job_tags=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in job_tags:
    job_title.append(i.text)
    
job_title=job_title[20:30]
```

```python
##SCRAPPING JOB LOCATION

loc_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in loc_tags:
    job_location.append(i.text)
    
    
job_location=job_location[:10]
```

```python
###SCRAPPING COMPANY NAME

company=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')
for i in company:
        company_name.append(i.text)
        
company_name=company_name[:10]
```

```python
###SCRAPPING EXPERIENCE REQUIRED

exp=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')

for i in exp:
    experience_required.append(i.text)
    
experience_required=experience_required[:10]
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
df=pd.DataFrame({'JObTitle':job_title,'JobLoacation':job_location,'CompanyName':company_name,'ExperienceRequired':experience_required})
df=df.head(10)
df
```

```python
name=[]
price=[]
```

```python
#scrapping name and price
nt=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
pt=driver.find_elements(By.XPATH,'//p/strong')
for i in nt:
    name.append(i.text)
    
for j in pt:
    price.append(j.text)
```

```python
name=name[:50]
price=price[:50]
```
