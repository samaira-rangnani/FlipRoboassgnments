```python
##seperating Features and label columns
x=df.drop('glasstype',axis=1)
y=df['glasstype']
```

```python
x.shape
```

```python
#Featuring Scaling on Independent variables as columns are not scaled
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
#Checking For multicollinearity using VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
vi=pd.DataFrame()
vi['Columns']=x.columns
vi['Vif values']=[vif(x.values,i) for i in range(len(x.columns))]
```

```python
vi
```

```python
cor['glasstype'].sort_values()
```

```python
#As columns Mg and Ca are showing multicollinearity between them,hence dropping Ca column
#as its corr value is less
x.drop('Ca',axis=1,inplace=True)
```

```python
#Checking for multi collinearity again
vi=pd.DataFrame()
vi['Columns']=x.columns
vi['Vif values']=[vif(x.values,i) for i in range(len(x.columns))]
```

```python
vi
```

```python
#As columns Al ,Na ,Si are showing multicollinearity 
#Dropping Si and Na as they have les correlation values between them
x.drop('Si',axis=1,inplace=True)
x.drop('Na',axis=1,inplace=True)
```

```python
x.shape
```

```python
y.value_counts()
```

```python
#we notice there is huge imbalance in target variable ,hence balancing using SMOTE technique
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
#Hence the classes are balanced
#From above observations made the problem is of Classification
```

```python
#Model building
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC

```

```python
maxacc=0
maxrs=0
for i in range(1,201):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.30)
    rfc=RandomForestClassifier()
    rfc.fit(x_train,y_train)
    ypred=rfc.predict(x_test)
    acc=accuracy_score(y_test,ypred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('Max accuracy score ',maxacc,'at random state',i)
```

```python
#Building train test split at maximum accuracy
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.30)
```

```python
#RFC
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
ypred=rfc.predict(x_test)
acc=accuracy_score(y_test,ypred)
print(acc)
print(confusion_matrix(y_test,ypred))
print(classification_report(y_test,ypred))
```

```python
#LR
lr=LogisticRegression()
lr.fit(x_train,y_train)
ypred=lr.predict(x_test)
acc=accuracy_score(y_test,ypred)
print(acc)
print(confusion_matrix(y_test,ypred))
print(classification_report(y_test,ypred))
```

```python
#SVC
svc=SVC()
svc.fit(x_train,y_train)
ypred=svc.predict(x_test)
acc=accuracy_score(y_test,ypred)
print(acc)
print(confusion_matrix(y_test,ypred))
print(classification_report(y_test,ypred))
```

```python
#BC
bc=BaggingClassifier()
bc.fit(x_train,y_train)
ypred=bc.predict(x_test)
acc=accuracy_score(y_test,ypred)
print(acc)
print(confusion_matrix(y_test,ypred))
print(classification_report(y_test,ypred))
```

```python
#GBC
gbc=GradientBoostingClassifier()
gbc.fit(x_train,y_train)
ypred=gbc.predict(x_test)
acc=accuracy_score(y_test,ypred)
print(acc)
print(confusion_matrix(y_test,ypred))
print(classification_report(y_test,ypred))
```

```python
#Above all models giving accuracy of 0.97
#Lets check cross_val_score
```

```python
#RFC
score=cross_val_score(rfc,x,y,cv=5)
print(score)
print(score.mean())
print('Difference between accuracy score and crossvalscore ',(acc-score.mean())*100)
```

```python
##checking Area Under Curve
from sklearn import metrics
fpr,tpr,thresholds=metrics.roc_curve(y_test,ypred)
roc_auc=metrics.auc(fpr,tpr)
display=metrics.RocCurveDisplay(fpr=fpr,tpr=tpr,roc_auc=roc_auc,estimator_name=svc)
display.plot()
```
