MODEL BUILDING

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor

from sklearn.model_selection import cross_val_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#LR model
lr=LinearRegression()
lr.fit(x_train,y_train)
ypred=lr.predict(x_test)
print(mean_squared_error(y_test,ypred))
print(mean_absolute_error(y_test,ypred))
print(r2_score(y_test,ypred))
```

```python
#Ridge
rd=Ridge()
rd.fit(x_train,y_train)
ypred=rd.predict(x_test)
print(mean_squared_error(y_test,ypred))
print(mean_absolute_error(y_test,ypred))
print(r2_score(y_test,ypred))
```

```python
##DTR
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
ypred=dtr.predict(x_test)
print(mean_squared_error(y_test,ypred))
print(mean_absolute_error(y_test,ypred))
print(r2_score(y_test,ypred))
```

```python
#SVR
svr=SVR()
svr.fit(x_train,y_train)
ypred=svr.predict(x_test)
print(mean_squared_error(y_test,ypred))
print(mean_absolute_error(y_test,ypred))
print(r2_score(y_test,ypred))
```

```python
#RFC
rfc=RandomForestRegressor()
rfc.fit(x_train,y_train)
ypred=rfc.predict(x_test)
print(mean_squared_error(y_test,ypred))
print(mean_absolute_error(y_test,ypred))
print(r2_score(y_test,ypred))
```

```python
#GBR
gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)
ypred=gbr.predict(x_test)
print(mean_squared_error(y_test,ypred))
print(mean_absolute_error(y_test,ypred))
print(r2_score(y_test,ypred))
```

```python
##From above  we observe that r2 score of Gradient boosting Regressor is good,with less error
#One again checking the cross_val_score
```

```python
#LR
score=cross_val_score(lr,x,y,cv=5)
print(score)
print(score.mean())
print('Difference between cross val score and r2 score is',(r2_score(y_test,ypred)-score.mean())*100)
```

```python
#Ridge
score=cross_val_score(rd,x,y,cv=5)
print(score)
print(score.mean())
print('Difference between cross val score and r2 score is',(r2_score(y_test,ypred)-score.mean())*100)
```

```python
#SVR
score=cross_val_score(svr,x,y,cv=5)
print(score)
print(score.mean())
print('Difference between cross val score and r2 score is',(r2_score(y_test,ypred)-score.mean())*100)
```

```python
#DTR
score=cross_val_score(dtr,x,y,cv=5)
print(score)
print(score.mean())
print('Difference between cross val score and r2 score is',(r2_score(y_test,ypred)-score.mean())*100)
```

```python
#RFC
score=cross_val_score(rfc,x,y,cv=5)
print(score)
print(score.mean())
print('Difference between cross val score and r2 score is',(r2_score(y_test,ypred)-score.mean())*100)
```

```python
#GBR
score=cross_val_score(gbr,x,y,cv=5)
print(score)
print(score.mean())
print('Difference between cross val score and r2 score is',(r2_score(y_test,ypred)-score.mean())*100)
```

```python
#From above cross-val-score ,we observe that Ridge Regression providing good accuracy with less error
```
