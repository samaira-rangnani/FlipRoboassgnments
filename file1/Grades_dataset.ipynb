```python
#Checking for VIF for confirming multicollinearity between columns
from statsmodels.stats.outliers_influence import variance_inflation_factor as vi
vif=pd.DataFrame()
vif['values']=[vi(x.values,i) for i in range(len(x.columns))]
vif['columns']=x.columns
```

```python
#As we observe that CY and MT column are high collinear dropping any 1 column
#As we observe that CS-214 and CS-306 column are high collinear dropping any 1 column
#As we observe that CS-317 and CS-412 column are high collinear dropping any 1 column
#As we observe that HS and EL column are high collinear dropping any 1 column
#CS-318 and CS-211 coolumns are highly collinear dropping any 1 column
#CS-312 and MT-224 columns are highly collinear dropping any 1 column
#CS-203 and TC-383 columns are highly collinear dropping any 1 column
#EE-119 and MT-222 columns are highly collinear dropping any 1 column
#HS-105/12 and CS-403 columns are highly collinear dropping any 1 column
#EL-102 and EE-222  columns are collinear dropping any 1 column
#MT-331 and CS-105 columns are collinear dropping any 1 column






x.drop('CY-105',axis=1,inplace=True)
x.drop('CS-214',axis=1,inplace=True)
x.drop('CS-421',axis=1,inplace=True)
x.drop('EL-332',axis=1,inplace=True)
x.drop('CS-318',axis=1,inplace=True)
x.drop('MT-224',axis=1,inplace=True)
x.drop('CS-203',axis=1,inplace=True)
x.drop('EE-119',axis=1,inplace=True)
x.drop('CS-403',axis=1,inplace=True)
x.drop('EE-222',axis=1,inplace=True)
x.drop('MT-331',axis=1,inplace=True)






```

```python
#As there are multiple columns hence uing PCA method which combines high collinear columns
#Here reducing multidimensionilty of columns from 29 to 15 columns
from sklearn.decomposition import PCA
pca=PCA(n_components=15)
x=pca.fit_transform(x)
x.shape
```

```python
x=pd.DataFrame(data=x)
x
```

```python
y
```

```python
#As we observe that traget variable y has continous and unique data,the problem is considered as Regression problem
```

```python
#Training the model

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import ExtraTreeRegressor
from sklearn.linear_model import Lasso,Ridge
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor
```

```python
#LR
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=45)

lr=LinearRegression()
lr.fit(x_train,y_train)
ypred=lr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
#lasso
ls=Lasso()
ls.fit(x_train,y_train)
ypred=ls.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
##Ridge
rd=Ridge()
rd.fit(x_train,y_train)
ypred=rd.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
##RFR
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
ypred=rfr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
##SVR
svr=SVR()
svr.fit(x_train,y_train)
ypred=svr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
##ETR
etr=ExtraTreeRegressor()
etr.fit(x_train,y_train)
ypred=etr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
##GBC
gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)
ypred=gbr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
##ABR
abr=AdaBoostRegressor()
abr.fit(x_train,y_train)
ypred=abr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
##from above we observe that LinearRegression and Ridge model are giving best accuracy with minimum error
#lets check cross val score for the models and decide to choose the highest accuracy model performance
```

```python
##lR
score=cross_val_score(lr,x,y,cv=5)
print(score)
print(score.mean())
print("Difference between accuracy and cross_val_score is",(r2_score(y_test,ypred)-score.mean())*100)
```

```python
##lasso
score=cross_val_score(ls,x,y,cv=5)
print(score)
print(score.mean())
print("Difference between accuracy and cross_val_score is",(r2_score(y_test,ypred)-score.mean())*100)
```

```python
##Ridge
score=cross_val_score(rd,x,y,cv=5)
print(score)
print(score.mean())
print("Difference between accuracy and cross_val_score is",(r2_score(y_test,ypred)-score.mean())*100)
```

```python
##RFR
score=cross_val_score(rfr,x,y,cv=5)
print(score)
print(score.mean())
print("Difference between accuracy and cross_val_score is",(r2_score(y_test,ypred)-score.mean())*100)
```

```python
##SVR
score=cross_val_score(svr,x,y,cv=5)
print(score)
print(score.mean())
print("Difference between accuracy and cross_val_score is",(r2_score(y_test,ypred)-score.mean())*100)
```

```python
##GBR
score=cross_val_score(gbr,x,y,cv=5)
print(score)
print(score.mean())
print("Difference between accuracy and cross_val_score is",(r2_score(y_test,ypred)-score.mean())*100)
```

```python
##ABR
score=cross_val_score(abr,x,y,cv=5)
print(score)
print(score.mean())
print("Difference between accuracy and cross_val_score is",(r2_score(y_test,ypred)-score.mean())*100)
```

```python
##ETR
score=cross_val_score(etr,x,y,cv=5)
print(score)
print(score.mean())
print("Difference between accuracy and cross_val_score is",(r2_score(y_test,ypred)-score.mean())*100)
```

```python
#From above cross validation score ,we observe that RandomForestRegressor getting high accuracy with minimum difference between error values
```

```python
#Hence choosing Random forest Regressor as final model
```

```python
#Hyper parameter tuning for RFR
from sklearn.model_selection import GridSearchCV

parameters={'n_estimators': [25, 50, 100, 150], 
    'max_features': ['sqrt', 'log2'], 
    'max_depth': [3, 6, 9], 
    'max_leaf_nodes': [3, 6, 9]}

gcv=GridSearchCV(RandomForestRegressor(),parameters)
gcv.fit(x_train,y_train)
gcv.best_params_
```

```python
rfr=RandomForestRegressor(max_depth= 9,
 max_leaf_nodes= 9,
 n_estimators= 100)
rfr.fit(x_train,y_train)
ypred=rfr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```
